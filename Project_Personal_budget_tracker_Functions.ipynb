```python
# File: project_personal_budget_tracker.py

# Step 1: Create empty lists
my_income_entries = []  # List to store all income entries
my_expense_entries = []  # List to store all expense entries

# Step 2: Function to add income
def add_income():
    description = input('Enter the description of the income: ')  # Ask user for income description
    amount = float(input('Enter the amount of income: '))  # Ask user for income amount and convert to float
    my_income_entries.append({'description': description, 'amount': amount})  # Store the entry as a dictionary
    print('Income added successfully.')  # Confirm entry was added

# Step 3: Function to add expense
def add_expense():
    description = input('Enter the description of the expense: ')  # Ask user for expense description
    amount = float(input('Enter the amount of expense: '))  # Ask user for expense amount and convert to float
    my_expense_entries.append({'description': description, 'amount': amount})  # Store the entry as a dictionary
    print('Expense added successfully.')  # Confirm entry was added

# Step 4: Function to view budget summary
def view_summary():
    total_income = sum(i['amount'] for i in my_income_entries)  # Sum of all income amounts
    total_expense = sum(i['amount'] for i in my_expense_entries)  # Sum of all expense amounts
    balance = total_income - total_expense  # Calculate remaining balance

    # Print the summary
    print('Your total income is:', total_income)
    print('Your total expense is:', total_expense)
    print('Your total balance is:', balance)

# Step 5: Function to delete an entry
def delete_entry():
    delete = input('What would you like to remove? (income/expense): ')  # Ask user what to delete
    if delete == 'income':
        des = input('Enter the description you would like to remove: ')  # Ask for description to remove
        for i in my_income_entries:
            if des == i['description']:  # If match found
                my_income_entries.remove(i)  # Remove it from the list
                print('Income item deleted successfully.')
                return
        print('Description not found.')  # If no match found
    elif delete == 'expense':
        des = input('Enter the description you would like to remove: ')  # Ask for description to remove
        for i in my_expense_entries:
            if des == i['description']:  # If match found
                my_expense_entries.remove(i)  # Remove it from the list
                print('Expense item deleted successfully.')
                return
        print('Description not found.')  # If no match found
    else:
        print('Invalid option. Please choose income or expense.')  # Handle wrong input


# Step 6: Function to modify an entry
def modify_entry():
    modify = input('What do you want to modify? (income/expense): ')  # Ask user which category to modify

    if modify == 'income':
        des = input('Enter the description you would like to modify: ')  # Ask which income item to change
        for i in my_income_entries:
            if des == i['description']:  # If matching income found
                new_description = input('Enter the new description: ')  # Get new description
                new_amount = float(input('Enter the new amount: '))  # Get new amount and convert to float
                i['description'] = new_description  # Update the description
                i['amount'] = new_amount  # Update the amount
                print('Income entry has been updated.')  # Confirm update
                return  # Stop the function once updated
        print('Description not found.')  # If no match was found

    elif modify == 'expense':
        des = input('Enter the description you would like to modify: ')  # Ask which expense item to change
        for i in my_expense_entries:
            if des == i['description']:  # If matching expense found
                new_description = input('Enter the new description: ')  # Get new description
                new_amount = float(input('Enter the new amount: '))  # Get new amount and convert to float
                i['description'] = new_description  # Update the description
                i['amount'] = new_amount  # Update the amount
                print('Expense entry has been updated.')  # Confirm update
                return  # Stop the function once updated
        print('Description not found.')  # If no match was found

    else:
        print('Invalid option. Please enter income or expense.')  # Handle incorrect input


# Step 7: Main menu to use the functions
while True:
    print('Personal Budget Tracker Menu:')  # Display main menu title
    print('1. Add income')
    print('2. Add expense')
    print('3. View summary')
    print('4. Delete an entry')
    print('5. Modify an entry')
    print('6. Exit')

    choice = input('Choose an option (1-6): ')  # Get user menu choice

    if choice == '1':
        add_income()  # Call function to add income
    elif choice == '2':
        add_expense()  # Call function to add expense
    elif choice == '3':
        view_summary()  # Call function to view income/expense summary
    elif choice == '4':
        delete_entry()  # Call function to delete an entry
    elif choice == '5':
        modify_entry()  # Call function to modify an entry
    elif choice == '6':
        print('Exiting Personal Budget Tracker. Goodbye!')  # Exit message
        break  # Stop the loop and exit program
    else:
        print('Invalid choice. Please select between 1-6.')  # Handle wrong menu input

```
