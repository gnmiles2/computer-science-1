```python

# i dont know what to do. i couldnt find the filrs for steps for this project in the google classroom.
# This class handles restaurant menu and orders
class Restaurant:
    def __init__(self):  # runs when we make a new Restaurant
        self.menu = []  # empty list to store dishes
        self.orders = []  # empty list to store orders

    def add_dish_to_menu(self, dish):  # add a dish to the menu
        self.menu.append(dish)  # use append to add the dish object

    def place_order(self, order):  # add a new order
        self.orders.append(order)

    def view_menu(self):  # print all dishes in the menu
        print("----- Restaurant Menu -----")
        for dish in self.menu:  # go through the menu
            print(dish)  # print each dish (calls __str__ of Dish)
        print("---------------------------\n")

    def view_orders(self):  # show all orders made
        print("===== All Orders =====")
        for order in self.orders:  # go through all orders
            order.view_order()  # print the details of the order
        print("======================\n")

# This class represents each dish on the menu
class Dish:
    def __init__(self, name, price, category):  # set name, price, type of dish
        self.name = name  # e.g. Spaghetti
        self.price = price  # e.g. 12.99
        self.category = category  # e.g. Main Course

    def __str__(self):  # controls how the dish prints
        return f"{self.name} - ${self.price:.2f} ({self.category})"

# This class stores customer info
class Customer:
    def __init__(self, name, email):  # set name and email
        self.name = name
        self.email = email

# This class handles individual orders
class Order:
    def __init__(self, customer):  # order linked to a customer
        self.customer = customer
        self.dishes = []  # list to store what the customer ordered

    def add_dish(self, dish):  # add a dish to the order
        self.dishes.append(dish)

    def view_order(self):  # show order details
        print(f"Order for {self.customer.name} ({self.customer.email}):")
        total = 0  # keep track of price
        for dish in self.dishes:  # go through each ordered dish
            print(f"  - {dish}")  # print the dish
            total += dish.price  # add price to total
        print(f"Total: ${total:.2f}\n")

# START TESTING THE PROGRAM
if __name__ == "__main__":
    my_restaurant = Restaurant()  # create a new restaurant

    # Create 3 dishes and add to menu
    dish1 = Dish("Spaghetti Bolognese", 12.99, "Main Course")
    dish2 = Dish("Tiramisu", 6.50, "Dessert")
    dish3 = Dish("Caesar Salad", 8.99, "Starter")
    my_restaurant.add_dish_to_menu(dish1)
    my_restaurant.add_dish_to_menu(dish2)
    my_restaurant.add_dish_to_menu(dish3)

    customer1 = Customer("Alice", "alice@example.com")  # create a customer
    my_restaurant.view_menu()  # show the menu
    my_restaurant.view_orders()  # show any orders (should be empty)

    # Create a new restaurant with different dishes
    my_restaurant = Restaurant()
    dish1 = Dish("Spaghetti Carbonara", 15.99, "Main Course")
    dish2 = Dish("Panacoutta", 6.50, "Dessert")
    dish3 = Dish("Mozzarella De Buffalla", 8.99, "Starter")
    my_restaurant.add_dish_to_menu(dish1)
    my_restaurant.add_dish_to_menu(dish2)
    my_restaurant.add_dish_to_menu(dish3)

    customer2 = Customer("Claudia", "Claudia@ef.com")  # new customer
    my_restaurant.view_menu()  # show new menu

    # Create an order for Claudia
    order1 = Order(customer2)
    order1.add_dish(dish1)  # add Carbonara
    order1.add_dish(dish2)  # add Panacoutta
    my_restaurant.place_order(order1)  # save the order

    my_restaurant.view_orders()  # show all orders
```
